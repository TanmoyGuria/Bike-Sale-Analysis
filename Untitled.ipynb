# Analysis by Country

```python
df['Country'].value_counts()
```

```python
df['Country'].value_counts().plot(kind='bar',figsize=(12,4),color='c')
plt.title("Sales In Each Country",fontsize=17)
plt.ylabel("Sales",fontsize=15)
plt.xlabel("Country",fontsize=15)
```

```python
cont = pd.pivot_table(df, values='Revenue', index=['Year'],
                    columns=['Country'], aggfunc=np.sum).reset_index()
cont
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

# Loop through the columns (countries) and plot a line for each one
for country in cont.columns[1:]:  # Skip the 'Year' column
    plt.plot(cont['Year'], cont[country], label=country)

plt.xlabel('Year')
plt.ylabel('Total Revenue(in million)')
plt.title('Revenue Trend by Country Over the Years')
plt.legend(loc='best')
plt.grid(False)

# Show the plot
plt.show()





```

# Sell track

## Male

```python
male=df.loc[df['Customer_Gender']=="M"]
male_bikes =male.loc[df['Product_Category']=="Bikes"]
#==
male_country = pd.pivot_table(male_bikes, values='Order_Quantity', index=['Year'],
                    columns=['Country'], aggfunc=np.sum).reset_index()
male_country
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

# Loop through the columns (countries) and plot a line for each one
for country in male_country.columns[1:]:  # Skip the 'Year' column
    plt.plot(male_country['Year'], male_country[country], label=country)

plt.xlabel('Year')
plt.ylabel('Total Order')
plt.title('Order Trend by Country Over the Years')
plt.legend(loc='best')
plt.grid(False)

# Show the plot
plt.show()
```

## Female
